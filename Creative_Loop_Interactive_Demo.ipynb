```python
"""
CREATIVE LOOP: OPEN WEB VS WALLED GARDEN (Versione per Google Colab/Jupyter)
-------------------------------------------------------------------------------
Autore: Massimiliano Gasparini (ORCID: 0009-0004-4216-4852)

Questo script monolitico contiene sia l'Engine Matematico (Lotka-Volterra/Gompertz)
sia l'ambiente di simulazione per dimostrare la prevenzione del MAD
attraverso l'applicazione della metrica termodinamica (Bit/Watt).
"""

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# ==========================================
# 1. IL MOTORE CINEMATICO
# ==========================================
class CreativeLoopEngine:
    def __init__(self, alpha=0.05, beta=0.0005, gamma=1.5, lambd=0.01, Kh=500.0, Ks=5.0, phi=0.5):
        self.alpha = alpha   # Rigenerazione Umana
        self.beta = beta     # Interferenza Autòfaga (Tossicità)
        self.gamma = gamma   # Efficienza Replicativa Sintetica
        self.lambd = lambd   # Inibizione da RLHF Umano
        self.Kh = Kh         # Capacità Portante Umana
        self.Ks = Ks         # Capacità Portante Sintetica
        self.phi = phi       # Iniezione di Omeostasi

    def system_equations(self, y, t):
        h, s = y
        s_safe = max(s, 1e-9)

        dh_dt = self.alpha * h * (1 - h / self.Kh) - self.beta * h * s + self.phi
        ds_dt = self.gamma * s * np.log(self.Ks / s_safe) - self.lambd * s * h

        return [dh_dt, ds_dt]

    def simulate(self, h_init, s_init, time_steps=2000, t_max=100):
        t = np.linspace(0, t_max, time_steps)
        y0 = [h_init, s_init]
        sol = odeint(self.system_equations, y0, t)
        return t, sol

# ==========================================
# 2. LO STRESS TEST COMPARATIVO
# ==========================================
# SCENARIO 1: OPEN WEB (Collasso MAD)
# Inquinamento tossico alto, IA non limitata, simulazione in token generici
open_web_engine = CreativeLoopEngine(
    alpha=0.05, beta=0.05, gamma=2.0, lambd=0.01, Kh=100.0, Ks=100.0, phi=0.5
)
t_web, sol_web = open_web_engine.simulate(h_init=10.0, s_init=5.0, t_max=100)
h_web, s_web = sol_web[:, 0], sol_web[:, 1]

# SCENARIO 2: SANDBOX TERMODINAMICA (Creative Loop)
# Filtro Bit/Watt attivo (beta abbattuto, Kh alto per ospitare entropia umana)
sandbox_engine = CreativeLoopEngine(
    alpha=0.05, beta=0.0005, gamma=1.5, lambd=0.01, Kh=500.0, Ks=5.0, phi=0.5
)
t_box, sol_box = sandbox_engine.simulate(h_init=10.0, s_init=0.1, t_max=100)
h_box, s_box = sol_box[:, 0], sol_box[:, 1]

# ==========================================
# 3. RENDERIZZAZIONE GRAFICA INLINE
# ==========================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Validazione Architetturale: Collasso (Open Web) vs Omeostasi (Walled Garden)', fontsize=15, fontweight='bold')

# Grafico 1: Open Web
ax1.plot(t_web, h_web, color='#000080', linewidth=2.5, label='Informazione Umana (h)')
ax1.plot(t_web, s_web, color='#B22222', linewidth=2.5, label='Informazione Sintetica (s)')
ax1.set_title('Scenario A: Open Web (Nessun Filtro)')
ax1.set_xlabel('Evoluzione Temporale')
ax1.set_ylabel('Volume Informazionale')
ax1.axhline(y=0, color='black', linewidth=1)
ax1.fill_between(t_web, 0, 100, color='red', alpha=0.05, label='Autofagia MAD Irreversibile')
ax1.legend()
ax1.grid(True, linestyle=':', alpha=0.7)

# Grafico 2: Sandbox
ax2.plot(t_box, h_box, color='#000080', linewidth=2.5, label='Densità Umana (h) [Bit/Watt]')
ax2.plot(t_box, s_box, color='#B22222', linewidth=2.5, label='Densità Sintetica (s) [Bit/Watt]')
ax2.set_title('Scenario B: Sandbox (Filtro Termodinamico Attivo)')
ax2.set_xlabel('Evoluzione Temporale')
ax2.set_ylabel('Densità Termodinamica')
ax2.fill_between(t_box, 0, 500, color='green', alpha=0.05, label='Omeostasi Simbiotica')
ax2.legend()
ax2.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show() # <-- Comando critico per Colab: renderizza l'immagine sotto la cella
```
